Importing modules and libraries

```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

Load and inspect data

```python
species = pd.read_csv("species_info.csv")
observations = pd.read_csv("observations.csv")
print(species.head())
print(observations.head())
print(species.conservation_status.unique())
```

Question 1: What is the distribution of conservation_status for animals?

```python
species["conservation_status"] = species["conservation_status"].replace(np.nan, "Not applicable")
species2 = species[species.conservation_status != "Not applicable"]
plt.subplot(1,2,1)
sns.countplot(x = "conservation_status", data = species)
plt.xlabel("Conservation Status")
plt.xticks(rotation = 30)
plt.subplot(1,2,2)
sns.countplot(x = "conservation_status", data = species2)
plt.xlabel("Conservation Status")
plt.xticks(rotation = 30)
plt.show()
```

```python
# So first of all, we have changed the Nan values in the species dataframe of this conservation status column into Not applicable
# And then we can see from the graph on the left that most of the recorded species fall in this category
# it appears that there are no endangered, threatened and in recovery species but if we zoom into these categories, we can see
# from the graph on the right that there are in fact a small number of species that fall into these categories.
```

Question 2: Are certain types of species more likely to be endangered?


```python
print(species.scientific_name[species.conservation_status == "Endangered"])
```

```python
endangered = species[species.conservation_status == "Endangered"]
sns.countplot(data = endangered, x = "category")
plt.title("The count of endangered species in each category")
plt.xticks(rotation = 30)
plt.show()
```

```python
# There are more endangered mammals than any other species, it's endangered species count is almost 50% higher than the second 
# on the list which is bird.
```

Question 3: Which species were spotted the most at each park?


```python
print(observations.park_name.unique())
park1 = observations[(observations.park_name == "Great Smoky Mountains National Park") & (observations.observations >=140)]
print("The max number of observations for park 1 is : {}".format(park1.observations.max()))
park2 = observations[(observations.park_name == "Yosemite National Park") & (observations.observations >=215)]
print("The max number of observations for park 2 is : {}".format(park2.observations.max()))
park3 = observations[(observations.park_name == "Bryce National Park") & (observations.observations >=165)]
print("The max number of observations for park 3 is : {}".format(park3.observations.max()))
park4 = observations[(observations.park_name == "Yellowstone National Park") & (observations.observations >=315)]
print("The max number of observations for park 4 is : {}".format(park4.observations.max()))
print(len(observations))
print(len(park1))
print(len(park2))
print(len(park3))
print(len(park4))
# sns.barplot(data = park1, x = "scientific_name", y = "observations")
# plt.xticks(rotation = 40)
# plt.show()
parks = [park1, park2, park3, park4]
for park in parks:
    plt.clf()
    sns.barplot(data = park, x = "scientific_name", y = "observations", order = park.sort_values("observations", ascending = False).scientific_name)
    plt.xticks(rotation = 40)
    plt.show()
```
